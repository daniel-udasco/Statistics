# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Import "current-power.csv" Dataset

```python
# load dataset
df = pd.read_csv(r'unprocessed/current-power.csv',
                 delimiter=",")
print(df)
```

# Summarize Stats

```python
df.describe()
```

# Covariance of current and power FROM current-power.csv

```python
df.cov()
```

# Scatter Plot

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

# Correlation Coefficient of current and power FROM current-power.csv

```python
df.corr()
```

# Scatter and Line Plot w/ Regression Line

```python
# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='#ff0000',
            alpha=0.6,
            label='Data points')
# line plot
plt.plot(x, line,
         color='#000000',
         label=f'Regression line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Import "air-quality.csv" Dataset

```python
df_air = pd.read_csv(r'unprocessed/air-quality.csv',
                     delimiter=",")
df_air
```

# Summarize Stats

```python
df_air.describe()
```

# Scatter and Line Plot w/ Regression Line (NOx vs Humidity Correlation)

```python
# regression line
x = df_air['NOx']
y = df_air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='#2C2C2D',
            alpha=0.2,
            label='Datapoints')
# line plot
plt.plot(x, line,
         color='#ff0000',
         label=f'Regression line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('CO (ug/cu.m)')
plt.legend()
plt.show()
```

# Scatter and Line Plot w/ Regression Line (Temp vs Humidity Correlation)

```python
# regression line
x = df_air['Temperature']
y = df_air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x + intercept

# scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x, y,
            color='#0000ff',
            alpha=0.2,
            label='Datapoints')
# line plot
plt.plot(x,line,
         color='#ff0000',
         label=f'Regression line (r = {r_value:.3f})')

plt.title('Temperature vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('Temperature (C)')
plt.legend()
plt.show()
```
