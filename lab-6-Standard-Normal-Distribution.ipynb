# Import "current-test.csv" Dataset

```python
df = pd.read_csv(r"unprocessed/current-test.csv",
                 delimiter=',',
                 index_col='Sample')
df
```

## Summary of "current-test.csv" dataframe

```python
df.info()
```

## Summary of "current-test.csv" statistics

```python
df.describe()
```

## Z-scores of "current-test.csv"

```python
df['Z-score'] = stats.zscore(df['Current'],ddof=1)
df
```

## Z-scores of "current-test.csv": Summary of statistics

```python
df['Z-score'].describe()
```

## Histogram of "current-test.csv"

```python
# histogram plot
plt.hist(df['Current'], bins=5, 
         alpha = 1,
         color='#fff000',
         edgecolor='#000000')

plt.title('Histogram of Current Values',
          font='Poppins',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

plt.show()
```

```python
# Normal distribution curve of "current-test.csv"
```

```python
# normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color="#909000",
         linewidth=2)

plt.title('Normal Distribution of Current Values',
          font='Poppins',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')

plt.show()
```

```python
# Histogram with Normal distribution curve Plot of "current-test.csv"
```

```python
plt.hist(df['Current'],
         bins=5,
         density=True,
         alpha = 1,
         color='#fff000',
         edgecolor='#000000')

plt.title('Normal Distribution of Current Values',
          font='Poppins',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

# normal distribution curve
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
         color='#ff0000',
         linewidth=2)

plt.show()
```

## Standard Normal Distribution N(0,1) of "current-test.csv"

```python
x_max = df['Z-score'].max()
x_min = df['Z-score'].min()
x = np.linspace(x_min,x_max,100)

p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
         color='#909000',
         linewidth=1)

plt.title('Standard Normal Distribution of Current Values N(0,1)',
          font='Poppins',
          fontsize=12)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')

plt.show()
```
